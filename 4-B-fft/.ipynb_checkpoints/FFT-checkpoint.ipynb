```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
from math import pi, cos, sin
%matplotlib inline

X_START = 0
X_END = 10
DISCRETIZATION_FREQUENCY = 10 # Points per segment of length 1.
MARKER_SIZE = 3
EPS = 1e-3

def find_power_of_2(num):
    cur_pow = 1
    while (abs(num) > cur_pow):
        cur_pow *= 2
    return cur_pow

def fft(x, inverse=False):
    """
    Returns the X_k array of DFT(x) (or inverse DFT(x)), 0 <= k <= N - 1.
    """
    N = len(x)
    
    new_N = find_power_of_2(N)
    x += [0] * (new_N - N)
    N = new_N
    
    if (N <= 1):
        return x[:]

    """
    X_k         = E_k + exp(+-2 * i * pi * k / N) * O_k
    X_{k + N/2} = E_k - exp(+-2 * i * pi * k / N) * O_k.
    
    E_k is actually a DFT of Even indexed values, and O_k is DFT of odds.
    
    exp(-2 * i * pi * k / N) = exp(0) * exp(-2 * i * pi / N) * ... * exp(...)  } k times.
    """
    
    x_even = fft(x[0::2], inverse)
    x_odd  = fft(x[1::2], inverse) 
    
    # k == 0
    cur_root = complex(1, 0)
    
    # exp(+-2 * i * pi / N) = cos(angle) + i sin(angle)
    angle     = 2 * pi / N * (1 if not inverse else -1)
    root_diff = complex(cos(angle), sin(angle))
    
    ans = [complex(0, 0)] * N
    for i in range(N // 2):
        ans[i]          = x_even[i] + cur_root * x_odd[i]
        ans[i + N // 2] = x_even[i] - cur_root * x_odd[i]
        
        if inverse:
            ans[i]          /= 2
            ans[i + N // 2] /= 2
        
        cur_root *= root_diff;
    
    return ans

def get_discrete_points(f, start, end, freq):
    points_count = (end - start) * freq
    x = [start + i  / freq for i in range(points_count)]
    y = [0] * points_count
    
    for i in range(points_count):
        y[i] = f(x[i])
        
    return x, y

def test(f, coefficients_deleted=[0], label="function"):
    TOTAL_POINTS_COUNT = (X_END - X_START) * DISCRETIZATION_FREQUENCY
    
    points_x, points_y = get_discrete_points(f, X_START, X_END, DISCRETIZATION_FREQUENCY)
    g = fft(points_y)
    
    plt.grid(True)
    series = [plt.scatter(points_x, fft(g[:int(len(g) * (1 - del_coefs_part))], True)[:TOTAL_POINTS_COUNT],
                      s = MARKER_SIZE)
          for del_coefs_part in coefficients_deleted]
     
    plt.figlegend(series, ("D = {}%".format(del_coefs_part * 100)
                           for del_coefs_part in coefficients_deleted), 5)
    plt.show()
```

$f(x) = x$:

```python
f = lambda x: x
test(f, [0, 1e-3, 1e-2, 0.1])
```

Как видно, даже при экстримально малых $D$ графики совсем не похожи.

$f(x) = x^2$:

```python
f = lambda x: x**2
test(f, [0, 1e-3, 1e-2, 0.1])
```

Всё так же плохо...

$f(x) = sin(x)$:

```python
f = lambda x: sin(x)
test(f, [0, 1e-3, 1e-2, 1e-1])
```

Теперь при $D = 10^{-3}$ графики довольно похожи, что обусловлено тем, что аппроксимация $sin(x)$ с помощью разложения его в сумму синусов даёт то, что у этих всех синусов коэффициент при $x$ одинаковый, только коэффициент перед самим синусом будет $\frac{1}{N}$, который довольно малый.

$f(x) = sin(x^2)$:

```python
f = lambda x: sin(x**2)

DISCRETIZATION_FREQUENCY = 200 # Предлагаю увеличить частоту дискретизации,
                               # чтобы было более отчетливо видно разницу между графиками
test(f, [0, 1e-4, 1e-3, 1e-2])

DISCRETIZATION_FREQUENCY = 10
```

Получили похожие результаты. Стоит заметить, однако, что сходство функций возрастает с аргументом, что связано с тем, что те последние функции которые мы удаляем были ответственны за график при малых $x$.

$f(x) = \frac{sin(x)}{x}$:

```python
f = lambda x: sin(x) / (x + EPS)

DISCRETIZATION_FREQUENCY = 200 # Предлагаю увеличить частоту дискретизации,
                               # чтобы было более отчетливо видно разницу между графиками
test(f, [0, 1e-4, 1e-3, 1e-2])

DISCRETIZATION_FREQUENCY = 10
```

Ну здесь всё плохо...
